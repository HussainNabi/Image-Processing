```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
###: Load an image
img = cv2.imread('image/frog.JPG')
cv2.imshow('Input', img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.waitKey()
```

```python
### Task#1: Convolve an image with a 7x7 box filter, and output the original and smoothed image side by side
x = 7
box = np.ones((x,x), np.float32) / (x*x)
Convolved = cv2.filter2D(img, -1, box)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(Convolved)
ax[1].set_title("7x7 Box Filter")
ax[1].axis("off")
plt.show()
```

```python
###Task#2: Using an image of your choice, apply a 5x5 gaussian filter to the image, then a 21x21 gaussian filter and show the three images side by side.

#gussian sigma
sigma=5
sigma2=21
# Apply Gaussian blur
blurred1 = cv2.GaussianBlur(img, 
(0, 0), sigmaX=sigma, sigmaY=sigma)
blurred2 = cv2.GaussianBlur(img, 
(0, 0), sigmaX=sigma2, sigmaY=sigma2)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(blurred1)
ax[1].set_title("5x5 blur")
ax[1].axis("off")
ax[2].imshow(blurred2)
ax[2].set_title("21x21 blur")
ax[2].axis("off")
plt.show()
```

```python
###Task#3: Using an image of your choice, apply a 3x3 Laplacian filter to sharpen the image. 
# Convert image to float32 for processing
image_float = img.astype(np.float32) / 255.0
laplacian = cv2.Laplacian(image_float, cv2.CV_32F)
sharpened = np.clip(image_float - laplacian, 0, 1)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(laplacian)
ax[1].set_title("laplacian")
ax[1].axis("off")
ax[2].imshow(sharpened)
ax[2].set_title("sharpened")
ax[2].axis("off")
plt.show()
```
