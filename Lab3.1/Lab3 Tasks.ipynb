```python
import cv2
import numpy as np
```

```python
###: Load an image
img = cv2.imread('./images/input.jpg', 0)
R, C = img.shape
cv2.imshow('Input image', img)
cv2.waitKey()
```

```python
### 1.1.	Increase the size of the image. 
x = 2 
y = 2
M = np.float32([[x, 0, 0], [0, y, 0], [0, 0, 1]])
scaled_img = cv2.warpPerspective(img, M, (C, R))
cv2.imshow('scaled_img', scaled_img)
cv2.waitKey()
```

```python
### 1.2.	Rotate the image by 120 degrees.
img_rotation = cv2.warpAffine(img, cv2.getRotationMatrix2D((C/2,R/2),120,1),(C,R))
                                                    ###((Location),degree,scale),(???)
cv2.imwrite('rotation_out.jpg', img_rotation)
cv2.imshow('rotated_img', img_rotation)
cv2.waitKey()
```

```python
### 1.3.	Perform Sheer operations over the image.
x = 0.5 
y = 0.5
M = np.float32([[1, x, 0], [y, 1, 0], [0, 0, 1]])
    ###[stretch, rotate,???],[shear, stretch,???]["?god rays?","?god rays?",shrink]
sheared_img = cv2.warpPerspective(img, M, (C, R))
                            ###(img, rotation matrix,(window sizes))
cv2.imshow('img', sheared_img)
cv2.waitKey()
```

```python
### 2.1.	Negative
negative_image = 256 - 1 -img
cv2.imshow('img', negative_image)
cv2.waitKey()
```

```python
### 2.1 Log
c = 45
log_transformed = c * (np.log(img + 1))
log_transformed = np.uint8(log_transformed)
cv2.imshow('img', log_transformed)
cv2.waitKey()
```

```python
### 2.3.    Power law
gamma = 2
gamma2=0.5
adjusted_image = np.power(img / 255.0, gamma) * 255.0
adjusted_image = adjusted_image.astype(np.uint8)
cv2.imshow('img', adjusted_image)
cv2.waitKey()
```
