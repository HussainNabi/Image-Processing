```python
import cv2
```

```python
###: Load an image
img = cv2.imread('./images/input.jpg', 0)
cv2.imshow('Input image', img)
cv2.waitKey()
```

```python
###convert a color image in grayscale mode.
gray_img = cv2.imread('images/input.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('Grayscale', gray_img)
cv2.waitKey()
```

```python
###Save an image. 
cv2.imwrite('images/output.jpg', gray_img)
```

```python
### read an image and want to convert it into GrayScal. 
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale image', gray_img)
cv2.waitKey()
```

```python
cv2.imwrite('images/output2.jpg', gray_img)
```
